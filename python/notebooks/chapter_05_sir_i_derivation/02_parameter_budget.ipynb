## $R_0$ decomposition

For the SIR_I model in the disease-free equilibrium ($S/N = 1$):

$$R_0 = c_I \cdot \beta \cdot \tau_R$$

This decomposes into three multiplicative contributions:
- **$c_I$**: contacts per infectious person per day
- **$\beta$**: transmission probability per contact
- **$\tau_R$**: mean infectious duration in days

Each parameter contributes a factor; if any one is zero, the disease cannot invade.

```python
c_I = params['c_I']
beta = params['beta']
tau_R = params['tau_R']

R0 = c_I * beta * tau_R
print(f"Baseline R_0 = c_I * beta * tau_R")
print(f"             = {c_I} * {beta} * {tau_R}")
print(f"             = {R0:.2f}")
print()

# Sensitivity decomposition: by Theorem 6.X, each parameter's normalized
# sensitivity index of R_0 equals its multiplicative role:
print("Closed-form normalized sensitivity indices (book Theorem 6.X):")
print(f"  S^{{R_0}}_{{c_I}}    = +1   (R_0 is linear in c_I)")
print(f"  S^{{R_0}}_{{beta}}   = +1   (R_0 is linear in beta)")
print(f"  S^{{R_0}}_{{tau_R}}  = +1   (R_0 is linear in tau_R)")
print(f"  S^{{R_0}}_{{c_S}}    = 0    (R_0 does not depend on c_S)")
print(f"  S^{{R_0}}_{{c_R}}    = 0    (R_0 does not depend on c_R)")
```

```python
# Visualize the parameter contributions on a logR_0 axis
fig, ax = plt.subplots(figsize=(7, 4))
labels = ['$c_I$', '$\\beta$', '$\\tau_R$', 'Combined']
values = [c_I, beta, tau_R, R0]
log_values = [np.log(v) if v > 0 else 0 for v in values]

colors = [BOOK_COLORS['primary'], BOOK_COLORS['exposed'], BOOK_COLORS['lambda_hat'], BOOK_COLORS['neutral']]

ax.bar(labels, log_values, color=colors, alpha=0.85)
ax.set_ylabel('$\\log$(value)')
ax.set_title('$R_0$ decomposition (log scale)')
ax.axhline(0, color='k', lw=0.5, alpha=0.3)
plt.tight_layout()
plt.show()
```
