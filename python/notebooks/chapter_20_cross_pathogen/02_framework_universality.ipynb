## What this notebook does

Demonstrates the book's central pedagogical claim: a single set of principles — the catalogue of essential considerations — applies to diseases with vastly different transmission modes, time scales, and intervention classes. The notebook produces a single comparative chart that places COVID, Dengue, and HIV on a common axis and shows where each case study sits along each consideration.

## Comparative summary across the three case studies

```python
from shared.parameters import baseline_chapter_20
p = baseline_chapter_20()
diseases = list(p['diseases'])
transmission = list(p['transmission_mode'])
time_scale = list(p['time_scale_years'])
headline_R0 = list(p['headline_R0'])

print(f'{"Disease":<10s} | {"Mode":<20s} | {"Time scale (yr)":<18s} | {"R_0":<5s}')
print('-'*65)
for d, m, t, r in zip(diseases, transmission, time_scale, headline_R0):
    print(f'{d:<10s} | {m:<20s} | {str(t)+" yr":<18s} | {r:<5.2f}')
```

## The framework's universal levers

```python
# Framework-level comparison: which consideration is the operationally dominant one for each disease
dominant = {
    'COVID-19': {
        'primary_consideration': 'C8/C9 (under-reporting + identifiability)',
        'second': 'C10 (generation time during early growth)',
        'main_intervention_lever': 'NPIs (contact reduction)',
    },
    'Dengue': {
        'primary_consideration': 'C13 (seasonal R_0(t) envelope)',
        'second': 'C11 (vector-control sensitivity ranking)',
        'main_intervention_lever': 'biting-rate reduction (S^R0_a = +1)',
    },
    'HIV': {
        'primary_consideration': 'C7 (mixing structure heterogeneity)',
        'second': 'C4/C12 (vertical pathway, treatment-as-prevention modifies effective R)',
        'main_intervention_lever': 'targeted prevention + ART scale-up',
    },
}
for d, info in dominant.items():
    print(f'{d}:')
    for k, v in info.items():
        print(f'  {k}: {v}')
    print()
```

## Visual summary

```python
fig, ax = plt.subplots(figsize=(10, 5))
x_log = np.log10(time_scale)
y = [3.5, 4.5, 2.0]   # arbitrary y-positions for layout
colors_d = [BOOK_COLORS['primary'], BOOK_COLORS['secondary'], BOOK_COLORS['highlight']]
for i, d in enumerate(diseases):
    ax.scatter(x_log[i], headline_R0[i], s=300, c=colors_d[i], edgecolors='black',
               linewidths=1.5, zorder=3)
    ax.annotate(f'{d}\n({transmission[i]})', (x_log[i], headline_R0[i]),
                xytext=(15, 5), textcoords='offset points', fontsize=10)
ax.set_xticks([np.log10(0.25), np.log10(1), np.log10(5), np.log10(30)])
ax.set_xticklabels(['3 mo', '1 yr', '5 yr', '30 yr'])
ax.set_xlabel('Time scale of dynamics')
ax.set_ylabel('Representative R_0')
ax.set_title('One framework, three disease profiles spanning 100x in time scale')
ax.set_ylim(0, 6)
ax.axhline(1, ls=':', color='gray', alpha=0.6)
ax.grid(True, alpha=0.3)
fig.tight_layout()
plt.show()

print('Closing observation:')
print('  - Each disease privileges a different consideration as primary')
print('  - The same compartmental + sensitivity + fitting machinery handles all three')
print('  - The framework\'s value is exactly its ability to surface which consideration matters where')
```
